# Exposure Time Calculator

This is a the exposure time calculator.

Remark that with a HxRG detector, you cannot change the exposure time. You can simply change the number of group or the number of ramp. 

You can also change the number of drop and/or frame per group. 

The `fetch_snr()` enables us you reach a target signal to noise ratio by scanning either the _ngroup_ parameter or the _nramp_ one. 

There, the snr is computed for a given wavelength band. See below.

## load a simulation

```python
import slicersim
```

```python
config = slicersim.iotools.get_config(scene='supernova.toml')
sim = slicersim.Simulation.from_config(config)
```

## Change whatever you like in the simulation

for instance a c=0.3 supernovae at z=1

```python
sim.update(target__redshift = 1, target__c=0.3)
```

## And check how many groups you would need

to reach a S/R=10 on in the restframe [4000, 7000] band

```python
ngroup, snr = sim.fetch_snr(10, free_parameter="ngroup", 
                           lbda_range= [4000, 7000], frame="rest")
print(f"with {ngroup} groups you reach a snr of {snr:.2f}")
```

```python
sim.update(ngroup=ngroup)
sim.get_times()
```
