```python
lbda, flux, variance = bd17.get_spectrum(unit="flambda", incl_error=True)
error = np.sqrt(variance)
```

```python
import matplotlib.pyplot as plt

fig, ax = plt.subplots()

ax.plot(lbda, flux, lw=1)
ax.fill_between(lbda, flux-error, flux+error, alpha=0.5)
```
