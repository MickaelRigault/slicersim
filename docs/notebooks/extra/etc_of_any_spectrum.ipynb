# Exposure time calculator 


***

## 0. Your spectrum

A spectrum should be an object with a wavelength (in angstrom) and a flux. 

You adapt the magnitude of input flux while loading the `LazuliTarget` in a given band. If not, the input flux will be assumed to be in physical units (f-lambda: erg/s/cm2/A).

For this example, let's assume a simple flat spectrum.

```python
import numpy as np
# define a point source
lbda_ref = np.linspace(3000, 20_000, 500) # larger than lazuli bandpass
flux_ref = np.ones(lbda_ref.shape) # flat spectrum
```

## 1. Load a LazuliTarget

Let's create a `LazuliTarget` with a magnitude of 20 in the bessellb-band.

```python
import slicersim
```

```python
target = slicersim.LazuliTarget(lbda_ref, flux_ref, mag=20, band="bessellb")
```

## 2. `setup_` the system

Let's say you want to reach an average signal-to-noise ratio of 20 within the [4000, 7000] angstrom wavelength range. 

```python
# set inplace=False to avoid setting target to the returned configuration.
config, reached_snr = target.setup_to_snr(20, lbda_range=[4000, 6800], frame='rest', 
                                          statistic=np.nanmean)
```

## 3. `get_` what you want.

```python
target.get_exposure_time()
```

```python
target.get_exposure_time(full=True)
```

## 4. `set_` properties of your target or `change_` your instrument

let's imagine that you want to 
- set the magnitude of the target to `mag=21` in `sdssr` band.
- change the read-out mode of your detector to have 4 frames per group with at most 80 groups per ramp

```python
target.set_properties(mag=21, band="sdssr")
```

```python
# here let's set 1 group, no drop, 4 frames per group. 
# The 1 group will be updated with setup_to_snr
target.change_detector_mode(nmd=(1,4,0), max_group=80)
```

```python
config, reached_snr = target.setup_to_snr(20)
config
```

```python
target.get_exposure_time()
```

## top-level `slicersim.lazuli_etc()`

The top-level function `slicersim.lazuli_etc()` runs all that for you at once.

```python
slicersim.lazuli_etc(lbda_ref, flux_ref, snr=20, 
                    mag=21, band="bessellb")
```
