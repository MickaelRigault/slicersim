# Exposure time calculator: Any spectrum

You can use `slicersim.Simulator` to compute the expected exposure time of any input spectrum. This makes use of `slicersim.PointSource.from_spectrum()`.



***

## 1. Generate a simulator

```python
import slicersim
from slicersim import iotools

# set the simulation
config = iotools.get_config()
sim = slicersim.Simulation.from_config(config)
```

## 2. Create a new point source

using the spectrum you want. Here a flat spectrum, and set mag=20 as default magnitude (band='bessellb' by default)

```python
import numpy as np

# define a point source
lbda_ref = np.linspace(3000, 20_000, 500)
flux_ref = np.ones(lbda_ref.shape) # flat spectrum

# and build the pointsource object
pointsource = slicersim.PointSource.from_spectrum(lbda_ref, flux_ref, 
                                                  mag=20 # default magnitude
                                                 )
```

## 3. Change the simulation to use this pointsource

```python
sim.change_target(pointsource)
```

***
## Run the snr measurement

Say here we want to vary the number of frame (this assumes single-frame groups)

```python
nframes, snr, exptime = sim.fetch_snr(10, free_parameter="nframe")
nframes, snr, exptime
```

Or the number of ramps

```python
sim.update(target__mag=25)
nramps, snr, exptime = sim.fetch_snr(10, free_parameter="nramp")
nramps, snr, exptime
```

Or the number of groups

```python
sim.update(target__mag=22)
ngroup, snr, exptime = sim.fetch_snr(10, free_parameter="ngroup")
ngroup, snr, exptime
```
